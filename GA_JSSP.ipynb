```python
import numpy as np
from random import choices
from JobShopScheduling import JSSP
import matplotlib.pyplot as plt
from GeneticAlgorithm import *

np.random.seed(0)
%load_ext autoreload
%autoreload 2
```

```python
model = JSSP(10, 10, Processing_time=0, randopt=True)
print(model.Processing_time)
```

```python
cross_methods = ['Uniform', 'OnePoint', 'TwoPoint', 'multi']
fits = []
for m in cross_methods:
    seq, stats = genetic_algorithm(model, max_iter=2000, pop_size=100, cross_method = m)
    print(seq)
    print("time:",stats[1])
    fits.append(stats[0])
```

```python
fig, ax = plt.subplots()
for i, m in enumerate(cross_methods):
    ax.plot(fits[i], label=m)
   
ax.legend()
ax.set_ylabel("makespan")
ax.set_xlabel("#iterations")
ax.set_title("Varied crossover methods for GA")
plt.show()
```
